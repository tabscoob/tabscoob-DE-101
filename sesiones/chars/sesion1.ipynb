## Dataframes

```python
df_credit = pd.read_csv('../../resources/credits.csv')
df_keywords = pd.read_csv('../../resources/keywords.csv')
df_links_small = pd.read_csv('../../resources/links_small.csv')
df_links = pd.read_csv('../../resources/links.csv')
df_ratings_small = pd.read_csv('../../resources/ratings_small.csv')
df_movies_metadata = pd.read_csv('../../resources/movies_metadata.csv')
```

## Cleaning df_movies_metadata['id']

```python
df_movies_metadata.head(5)
```

```python
df_movies_metadata['id'] = pd.to_numeric(df_movies_metadata['id'], errors='coerce', downcast='integer')
df_movies_metadata.shape
```

```python
df_movies_metadata.dropna(subset=['id'], inplace=True)
df_movies_metadata.shape
```

```python
df_movies_metadata['id'].astype('int64')
```

## Getting top 10 raiting movies

```python
df_ratings_small.sort_values('rating', ascending=False)
```

```python
df_top_ten = df_ratings_small.sort_values('rating', ascending=False).head(10)
df_top_ten
```

## Merge df_ratings with df_movies_metadata

```python
df_topmovies_rating = pd.merge(df_top_ten, df_movies_metadata, how='inner', left_on='movieId', right_on='id')
df_topmovies_rating[['movieId','original_title', 'vote_average', 'rating']].sort_values('vote_average', ascending=False)
```
